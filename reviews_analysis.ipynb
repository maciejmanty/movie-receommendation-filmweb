```python
df['user_id'] = df.groupby('user_name').ngroup()
df['movie_id'] = df.groupby('movie').ngroup()
```

```python
df.sort_values('user_id')
```

```python
userRatings = df.pivot_table(index=['user_id'],columns=['movie'],values='score')
userRatings.head()
```

```python
corrMatrix = userRatings.corr()
corrMatrix.head()
```

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=10)
corrMatrix.head()
```

```python
df['user_id'].value_counts().head(20)
```

```python
myRatings = userRatings.loc[11771].dropna()
myRatings
```

```python
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    print ("Adding sims for " + myRatings.index[i] + "...")
    # Retrieve similar movies to this one that I rated
    sims = corrMatrix[myRatings.index[i]].dropna()
    # Now scale its similarity by how well I rated this movie
    sims = sims.map(lambda x: x * myRatings[i])
    # Add the score to the list of similarity candidates
    simCandidates = simCandidates.append(sims)
    
#Glance at our results so far:
print ("sorting...")
simCandidates.sort_values(inplace = True, ascending = False)
print (simCandidates.head(10))
```

```python
simCandidates = simCandidates.groupby(simCandidates.index).sum()
```

```python
simCandidates.sort_values(inplace = True, ascending = False)
simCandidates.head(10)
```

```python
filteredSims = simCandidates.drop(myRatings.index)
filteredSims.head(10)
```
